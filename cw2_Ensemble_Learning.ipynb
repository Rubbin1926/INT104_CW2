```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

data = pd.read_excel("CW_Data.xlsx")

# 移除索引列
data = data.drop('Index', axis=1)

data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = data.drop(['Programme'], axis=1)
y = data['Programme']

pca = PCA(n_components=2)
X = StandardScaler().fit_transform(X.values)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2']).values
print(principalDf.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(principalDf, y.values.tolist(), test_size=0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Visualization(clf, x_train, x_test, y_train, y_test):
    # Plotting sample scatter plot
    plt.scatter(x_train[:,0], x_train[:,1], c=y_train, zorder=10, cmap=plt.cm.Paired,
                edgecolor='k', s=20)
    # Plotting decision boundary, selecting minimum and maximum values
    x_min = x_train[:, 0].min()
    x_max = x_train[:, 0].max()
    y_min = x_train[:, 1].min()
    y_max = x_train[:, 1].max()
    # Generating grid matrix
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    # Calculating the predicted class for each grid point
    Z = clf.predict(np.c_[XX.ravel(), YY.ravel()])
    # Reshaping the predicted class to match the grid shape
    Z = Z.reshape(XX.shape)
    # Plotting the decision regions
    plt.pcolormesh(XX, YY, Z, cmap=plt.cm.Paired)
    # Plotting the test set scatter plot
    plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap=plt.cm.Paired,
                edgecolor='k', s=10)
    plt.title('Accuracy:{}'.format(clf.score(x_test, y_test)))
```

```python
# SVM
from sklearn import svm
svm_clf = svm.SVC(kernel='poly', degree=3, coef0=1, C=5)
svm_clf.fit(x_train, y_train)
svm_score = svm_clf.score(x_test, y_test)
Visualization(svm_clf, x_train, x_test, y_train, y_test)
```

```python
# Decision Tree
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

DT_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, random_state=42)
DT_clf.fit(x_train, y_train)
DT_score = DT_clf.score(x_test, y_test)
Visualization(DT_clf, x_train, x_test, y_train, y_test)
```

```python
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
gnb_clf = GaussianNB()
gnb_clf.fit(x_train, y_train)
gnb_score = gnb_clf.score(x_test, y_test)
Visualization(gnb_clf, x_train, x_test, y_train, y_test)
```

```python
from sklearn.ensemble import VotingClassifier
classifier1 = svm_clf
classifier2 = DT_clf
classifier3 = gnb_clf

estimators = [('classifier1', classifier1), ('classifier2', classifier2), ('classifier3', classifier3)]
voting = VotingClassifier(estimators, voting='hard')

voting.fit(x_train, y_train)
Visualization(voting, x_train, x_test, y_train, y_test)
```
